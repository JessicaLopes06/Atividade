```python
# Tratar a coluna tempo
df['TEMPO'] = pd.to_datetime(df['TEMPO'])
df
```

```python
df.groupby('TEMPO').nunique()['ID'].plot()
```

```python
df.STATUS.value_counts()
```

```python
df['Mes'] = df['TEMPO'].dt.strftime('%b')
```

```python
# Defina a ordem dos meses em inglês
ordem_meses = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
# Converta a coluna 'Mes' para uma categoria com a ordem desejada
df['Mes'] = pd.Categorical(df['Mes'], categories=ordem_meses, ordered=True)

# Agrupe por mês e status, contando o número de ocorrências
agrupado = df.groupby(['Mes', 'STATUS']).size().unstack().fillna(0)

# Crie um gráfico de barras empilhadas
ax = agrupado.plot(kind='bar', stacked=True, figsize=(10, 6))

# Adicione rótulos de dados nas barras
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height:.0f}', (x + width/2, y + height/2), ha='center', va='center')

# Adicione rótulos e título ao gráfico
plt.xlabel('Mês')
plt.ylabel('Quantidade de Status')
plt.title('Status Mensal')

# Exiba o gráfico
plt.show()
```

```python
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
#Temas de Reclamações
dtemas = df.dropna(subset=["TEMA"],axis = 0)["TEMA"]
```

```python
#Nuvem de palavras dos temas da reclamação
#concatenar todo o texto
todo_texto = " ".join(s for s in dtemas )
todo_texto
```

```python
stopwords = set(STOPWORDS)
stopwords.update(['da','meu','está','e','o','de','para','em','A','E','I','O','U','um','as','ao','o','do','com','para','não','NÃO','aqui','que','eu','HÁ','pois','Hapvida','NAO','não','nao','fazer','foi','na','no','pelo','por','par','para','DESDE','os','data','tão','deu', 'estão','mas','já','até','Dr', 'sendo','dela','agora','vai','q','ter','esse','ano','toda','isso','já'])
```

```python
#Gerando a nuvem
wordcloud = WordCloud(stopwords=stopwords,
                      width = 1600, height = 800).generate(todo_texto)
```

```python
#mostrar nuvem
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation = 'bilinear')
ax.set_axis_off()
plt.imshow(wordcloud)
```

```python
##Descrição
ddesc = df.dropna(subset=["DESCRICAO"],axis = 0)["DESCRICAO"]
#concatenar todo o texto
DESC = " ".join(s for s in ddesc )
#Gerando a nuvem
wordcloud2 = WordCloud(stopwords=stopwords,
                      width = 1600, height = 800).generate(DESC)
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud2, interpolation = 'bilinear')
ax.set_axis_off()
plt.imshow(wordcloud2)
```
